```python
import pandas as pd
from tabpfn import TabPFNClassifier, TabPFNRegressor
from tabpfn_extensions.embedding import TabPFNEmbedding
import numpy as np
from dotenv import load_dotenv
from sklearn.model_selection import StratifiedShuffleSplit
load_dotenv(".env")
```

```python
blood_dataset = pd.read_csv("datasets/blood_dataset/blood.csv")
print(blood_dataset.head())
print(len(blood_dataset))
```

```python
from sklearn.metrics import accuracy_score
data = np.array(blood_dataset.drop(columns=["Class"]),dtype=np.float32)
targets = np.array(blood_dataset["Class"],dtype=np.float32)
sss = StratifiedShuffleSplit(1,test_size=0.2)
for i, (train_index, test_index) in enumerate(sss.split(data, targets)):
    X_train = data[train_index]
    y_train = targets[train_index]
    X_test = data[test_index]
    y_test = targets[test_index]
    embedding = TabPFNEmbedding()
    train_embeds = embedding.fit_transform(X_train, y_train)  # OOF embeddings
    test_embeds = embedding.transform(X_test)
    print(train_embeds.shape)
    print(test_embeds.shape)
    clf = TabPFNClassifier()
    clf.fit(X_train, y_train)  # downloads checkpoint on first use
    predictions = clf.predict(X_test)
    print(accuracy_score(y_test,predictions))
```
